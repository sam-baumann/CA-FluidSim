```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('timings.csv')
data.head()
```

```python
#create a new col by multiplying cell_count_x and cell_count_y
data['cell_count'] = data['cell_count_x'] * data['cell_count_y']
data.head()
```

```python
%matplotlib inline
#plot a line with the cell_count on the x-axis and the fps on the y-axis
plt.plot(data['cell_count'], data['fps'])
plt.xlabel('Number of Cells in the Grid')
plt.ylabel('Frames per Second (more is better)')
plt.title('Frames per Second vs. Number of Cells in the Grid')
#draw a horizontal line at 12fps
```

```python
%matplotlib inline
#the same plot but log scale on the y-axis
plt.plot(data['cell_count'], data['fps'])
plt.xlabel('Number of Cells in the Grid')
plt.ylabel('Frames per Second (more is better)')
plt.title('Frames per Second vs. Number of Cells in the Grid (log scale)')
#add a line at 12fps
plt.axhline(y=12, color='r', linestyle='-')
#fix the y ticks
plt.legend(['Observed frames per Second', '12fps'])
plt.yscale('log')
#fix the y ticks
plt.yticks([8, 12, 16, 32, 64, 128, 256, 512], [8, 12, 16, 32, 64, 128, 256, 512])
```
